```python
import numpy as np
import numpy.linalg as nplin
import matplotlib.pyplot as plt
from matplotlib import cm
from numpy.polynomial.polynomial import polyfit
```

# Number of nodes
## All experiments with nx = 520 ny=180
## MaxIter = 20 000

```python
# DATA

felix8  = np.array([3.84232990e-02, 2.37367026e+01, 4.75377468e+01, 7.08105984e+01, 9.48140632e+01, 1.18385816e+02, 1.42047015e+02, 1.66140285e+02, 1.90419284e+02, 2.15018669e+02, 2.39000450e+02, 2.62338231e+02, 2.85915972e+02, 3.09606183e+02, 3.33244863e+02, 3.56802872e+02, 3.80225792e+02, 4.03864918e+02, 4.27389398e+02, 4.51194381e+02])        
                 
felixb  = np.array([5.28140720e-02, 4.49787182e+01, 9.00108838e+01, 1.34334005e+02, 1.78751989e+02, 2.23177506e+02, 2.67463509e+02, 3.11891127e+02, 3.56440017e+02, 4.00868095e+02,4.45729241e+02, 4.90445497e+02, 5.35115496e+02, 5.79556377e+02, 6.24298392e+02, 6.69176946e+02, 7.14275416e+02, 7.58969280e+02, 8.03633584e+02, 8.48439443e+02]) 

felix16 = np.array([4.47635540e-02, 2.31385354e+01, 4.61257619e+01, 6.90754086e+01,9.19197540e+01, 1.14801914e+02, 1.37694728e+02, 1.60608704e+02,1.83525062e+02, 2.06396812e+02,2.29303615e+02, 2.52336850e+02,2.75295007e+02, 2.98212130e+02,3.21191395e+02, 3.44130342e+02, 3.67051482e+02, 3.89832456e+02,4.12729179e+02, 4.35510282e+02,]) 

felixcomp = felixb -felix16

grid = np.arange(0, 19000+1, 1000)
```

```python
# Applying a linear fit with polyfit
a, l = polyfit(grid, felixb, 1)
b, m = polyfit(grid, felix8, 1)
c, n = polyfit(grid, felix16, 1)
plt.xlabel('iterations')
plt.ylabel('time (s)')
plt.plot(grid, felixb, '.', label='serial code')
plt.plot(grid, felix8, '.', label='8 cores parallelisation')
plt.plot(grid, felix16, '.', label='16 cores parallelisation')
plt.plot(grid, a + l * grid, '-')
plt.plot(grid, b + m * grid, '-')
plt.plot(grid, c + n * grid, '-')
plt.legend(bbox_to_anchor=(0.1, 0.8), loc=6, borderaxespad=0.)
plt.savefig('compareIter0.png', bbox_inches='tight', dpi=200)
```

```python
plt.xlabel('iterations')
plt.ylabel('time (s)')
w, z = polyfit(grid, felix8-felix16, 1)
plt.plot(grid, felix8-felix16, '.', label='8 - 16 cores')
plt.plot(grid, w+ z * grid, '-')
plt.legend(bbox_to_anchor=(0.3, 0.9), loc=1, borderaxespad=0.)
plt.savefig('compareIter1.png', bbox_inches='tight', dpi=200)
print(z)
```

```python
plt.xlabel('iterations')
plt.ylabel('time (s)')
wc, zc = polyfit(grid, felixcomp, 1)
plt.plot(grid, felixcomp, '.', label='serial - 16 cores')
plt.plot(grid, wc+ zc * grid, '-')
plt.legend(bbox_to_anchor=(0.4, 0.9), loc=1, borderaxespad=0.)
plt.savefig('compareIter2.png', bbox_inches='tight', dpi=200)
print(zc)
```

```python
# DATA

nxny = np.array([  3744,  23400,  45864,  93600, 158184, 210600, 303264, 374400.])

#mpiduration16a = np.array([ 0.75636524,  2.38380766,  3.93047871,  7.12159491, 11.48587032, 15.63934277,21.85900711, 28.87241917])
#mpiduration8a =  np.array([ 0.64035288,  2.15838093,  3.78569864,  7.1380981,  12.60445388, 16.97447975, 24.32682124, 29.61884926])
#mpiduration8b =  np.array([ 0.68441827,  2.30965382,  3.95485697,  7.42331982, 12.93176608, 17.83714703, 25.49976937, 30.71541884])
#mpiduration16b =  np.array([ 0.81015234,  2.49433641,  4.07367647,  7.31719542, 11.75231774, 16.10256564, 22.86070271, 29.10477829] )
#mpiduration8 = (mpiduration8a+mpiduration8b)/2
#mpiduration16 = (mpiduration16a+mpiduration16b)/2
#duration =  np.array([ 0.60581661,  3.62989023,  7.24447179, 15.39110067, 23.59210151, 38.31227233, 56.18519771, 71.11993781])

duration = np.array(  [ 0.83745882,  6.20569589, 12.27412638, 24.80171968, 36.81331004, 58.87162254, 81.20444099, 98.96116806] ) 

mpiduration8 =  np.array([ 1.16754185,  3.85096114,  6.61481334, 12.50601114, 21.73988265, 29.96679801, 42.14666405, 51.84048606])

mpiduration16 =  np.array(  [1.36366659,  4.21110405,  6.86597792, 12.23344191, 19.64151567, 26.87309634, 37.88806279, 49.30170318] )
```

```python
# Applying a linear fit with polyfit
aa, ll = polyfit(nxny, duration, 1)
bb, mm = polyfit(nxny, mpiduration8, 1)
cc, nn = polyfit(nxny, mpiduration16, 1)
plt.xlabel('grid size')
plt.ylabel('time (s)')
plt.plot(nxny, duration, '.', label='serial code')
plt.plot(nxny, mpiduration8, '.', label='8 cores parallelisation')
plt.plot(nxny, mpiduration16, '.', label='16 cores parallelisation')
plt.plot(nxny, aa + ll * nxny, '-')
plt.plot(nxny, bb + mm * nxny, '-')
plt.plot(nxny, cc + nn * nxny, '-')
plt.legend(bbox_to_anchor=(0.1, 0.8), loc=6, borderaxespad=0.)
plt.savefig('compareSize0.png', bbox_inches='tight', dpi=200)
```

```python
plt.xlabel('grid size')
plt.ylabel('time (s)')
ww, yy = polyfit(nxny, mpiduration8-mpiduration16, 1)
plt.plot(nxny, mpiduration8-mpiduration16, '.', label='8 - 16 cores')
plt.plot(nxny, yy * nxny + ww, '-')
plt.legend(bbox_to_anchor=(0.3, 0.9), loc=1, borderaxespad=0.)
plt.savefig('compareSize1.png', bbox_inches='tight', dpi=200)
```

```python
servpara = duration - mpiduration16
plt.xlabel('grid size')
plt.ylabel('time (s)')
uu, ii = polyfit(nxny, servpara, 1)
plt.plot(nxny, servpara, '.', label='serial - 16 cores')
plt.plot(nxny, ii * nxny + uu, '-')
plt.legend(bbox_to_anchor=(0.4, 0.9), loc=1, borderaxespad=0.)
plt.savefig('compareSize2.png', bbox_inches='tight', dpi=200)
```
